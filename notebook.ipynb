Task: Geographic Analysis

Plot the locations of restaurants on a

```python
plt.figure(figsize=(10, 8))
plt.scatter(df['Longitude'], df['Latitude'], alpha=0.5, marker='.')
plt.title('Locations of Restaurants')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

```python
!pip install scikit-learn
```

```python
from sklearn.cluster import KMeans
```

```python
X = df[['Longitude', 'Latitude']]

# Determine the optimal number of clusters using the elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')  # Within cluster sum of squares
plt.show()

```

```python
# Based on the elbow curve, choose the optimal number of clusters and perform clustering
optimal_k = 2  # Choose the number of clusters based on the elbow method
kmeans = KMeans(n_clusters=optimal_k, init='k-means++', random_state=42)
kmeans.fit(X)

# Add cluster labels to the dataset
df['Cluster'] = kmeans.labels_

# Plot the clusters on the map
plt.figure(figsize=(10, 8))
for cluster in range(optimal_k):
    cluster_data = df[df['Cluster'] == cluster]
    plt.scatter(cluster_data['Longitude'], cluster_data['Latitude'], label=f'Cluster {cluster + 1}', alpha=0.5)
plt.title('Clusters of Restaurants')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Identify if there are any restaurant chains present in the dataset
restaurant_chains = df['Restaurant Name'].value_counts()
chains_present = restaurant_chains[restaurant_chains > 1].index.tolist()
print("Restaurant chains present in the dataset:",)
for i in chains_present:
    print(i)
```

```python
# Analyze the ratings and popularity of different restaurant chains
chain_ratings = df.groupby('Restaurant Name')['Aggregate rating'].mean()
chain_popularity = df.groupby('Restaurant Name')['Votes'].sum()

# Print average ratings and popularity of restaurant chains
for chain in chains_present:
    avg_rating = chain_ratings.get(chain, 0)
    total_votes = chain_popularity.get(chain, 0)
    print(f"Restaurant Chain: {chain} | Average Rating: {avg_rating:.2f} | Total Votes: {total_votes}")
```
